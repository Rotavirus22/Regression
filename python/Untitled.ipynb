## Calculating the R-Squared

```python
reg.score(x,y)
```

### Formula for the adjusted R^2
$R^2_{adj.} = 1 -(1-R^2)*\frac{n-1}{n-p-1}$

```python
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
adjusted_r2 =  1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```
